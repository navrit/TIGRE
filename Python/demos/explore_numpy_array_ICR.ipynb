```python
final_badmap = np.zeros_like(arr)
for slice in range(0,arr.shape[0]):
    current_proj = arr1[slice] - arr[slice]
    current_proj[current_proj == np.nan] = np.nan
    current_proj[current_proj == np.Inf] = np.nan
    current_proj[current_proj == -np.Inf] = np.nan
    mean_current_proj = np.nanmean(current_proj)
    std_current_proj = np.nanstd(current_proj)
    thresh = 3*std_current_proj

    current_proj[current_proj > mean_current_proj + thresh] = np.nan
    current_proj[current_proj < mean_current_proj - thresh] = np.nan

    current_badmap = np.zeros_like(current_proj)
    current_badmap *= current_proj
    current_badmap = np.nan_to_num(current_badmap,nan = 1)

    final_badmap[slice,:,:] = current_badmap
```

```python
final_bin = np.sum(final_badmap,axis=0)
final_bin[final_bin >= 1] = 1
print(final_bin.shape)
plt.imshow(final_bin, cmap=cm, interpolation='none')
plt.colorbar()
plt.show()
```

```python
from turtle import color

slice = 0
plt.close('all')

plt.figure(2)
im = plt.imshow(arr[:,slice,:], cmap=cm, interpolation='none')
plt.title('(x,z)')
im.cmap.set_under('b')
im.cmap.set_over('r')
im.cmap.set_bad('pink')
plt.colorbar()
plt.show()

current_row = arr[200,slice,:]
moving_window = 3
average_y = []
for ind in range(len(current_row) - moving_window + 1):
    average_y.append(np.median(current_row[ind:ind+moving_window]))

diff_res = average_y - current_row[:-moving_window+1]

badmap = np.ones((len(diff_res)))
for i in range(0,len(badmap)):
    current_perc = np.abs(((average_y[i] - current_row[i])/current_row[i])*100)
    if current_perc >= 10:
        badmap[i] = 1
        print(f'i:{i}, current_percentage: {current_perc}')
    else:
        badmap[i] = 0

mean_abs = np.mean(np.abs(diff_res)) + 2*np.std(np.abs(diff_res))

plt.figure(500)
plt.plot(np.abs(diff_res),'k')
plt.axhline(mean_abs)
plt.figure(534)
plt.plot(current_row)
plt.plot(average_y,'r')
plt.plot(badmap*current_row[:-moving_window+1])

plt.figure(101)
plt.hist(np.abs(diff_res),bins=100)
plt.show()
```
